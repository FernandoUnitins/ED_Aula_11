```python
class Node:
    def __init__(self, qtd, preco, descricao):
        self.qtd = qtd
        self.preco = preco
        self.descricao = descricao
        self.next = None

class OrdenedList:
    def __init__(self):
        self.head = None

    # Consulta se a lista está fazia
    def is_empty(self):
        return self.head is None

    # Inserção de elementos na lista
    def insert(self, qtd, preco, descricao):
        new_node = Node(qtd, preco, descricao)

        if self.is_empty() or descricao.lower() <= self.head.descricao.lower():
            new_node.next = self.head
            self.head = new_node
        else:
            current = self.head
            while current.next is not None and current.next.descricao.lower() < descricao.lower():
                current = current.next
            new_node.next = current.next
            current.next = new_node

    # Busca de elemento por descrição (usei o lower() para não diferenciar maiúsculas de minúsculas)
    def search(self, descricao):
        current = self.head
        while current is not None and current.descricao.lower() <= descricao.lower():
            if current.descricao.lower() == descricao.lower():
                return current
            current = current.next
        return None

    # Remoção de elemento por descrição (usei o lower() para não diferenciar maiúsculas de minúsculas)
    def remove(self, descricao):
        if self.is_empty():
            return None

        if self.head.descricao.lower() == descricao.lower():
            removed_node = self.head
            self.head = self.head.next
            return removed_node

        current = self.head
        while current.next is not None and current.next.descricao.lower() < descricao.lower():
            current = current.next

        if current.next is None or current.next.descricao.lower() > descricao.lower():
            return None

        removed_node = current.next
        current.next = current.next.next
        return removed_node

    # Exibição da lista - Display
    def display(self):
        if self.is_empty():
            print("A lista ordenada está vazia!")
        else:
            current = self.head
            while current:
                print(f"Qtd: {current.qtd}, Preço: {current.preco}, Descrição: {current.descricao}")
                current = current.next
```

```python
ol = OrdenedList()

# Inserção fora de ordem alfabética proposital
ol.insert(1, 10.0, "Fernando")
ol.insert(2, 5.0, "Aline")
ol.insert(3, 7.5, "Zezinho")
ol.insert(4, 8.0, "Snoopy")

print("Lista após inserções:")
ol.display()

# Testando busca
resultado = ol.search("Fernando")
if resultado:
    print(f"\nBusca: {resultado.descricao} encontrado com preço {resultado.preco}")
else:
    print("\nBusca: Descrição não encontrada.")

# Testando remoção
ol.remove("Aline")
print("\nLista após remoção:")
ol.display()
```
